```python
driver = webdriver.Chrome('chromedriver_linux64/chromedriver')
driver.get("https://www.vogue.in/fashion")
```

```python
def get_link():
    soup = BeautifulSoup(driver.page_source)
    for i in soup.findAll('div',{'class': 'Layout-cEduSl Wrapper-jAfsdL gZoiz'}):
        for items in i('a',{'data-test-id' : 'Anchor'}):
            link1 = "https://www.vogue.in" + items.get('href')
            #print(link1)
            link.append(link1)
```

```python
link = []
get_link()
for i in range(2,311):
    page = "https://www.vogue.in/fashion?page=" + str(i)
    print(page)
    driver.get(page)
    get_link()
    print(".....................")
print(link)
```

```python
csv_file = open('Vogue.csv', 'w',encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Title','Article Content'])
count = 0
for i in link: 
    driver.get(i)
    soup = BeautifulSoup(driver.page_source)
    title = soup.find('h1').text
    print(title)
    print("............................")
    content = ""
    for j in soup.findAll('p'):
        content1= j.text
        content = content +" "+ content1
    csv_writer.writerow([title, content])
    #print(content)
    print(count)
    count = count + 1
    print("............................")
```
