```python
driver = webdriver.Chrome('chromedriver_linux64/chromedriver')
driver.get("https://fashionista.com/fashion-week")
```

```python
link = []
soup = BeautifulSoup(driver.page_source)
for i in soup.findAll('section',{'class': 'm-list-hub m-component-stack mm-component-stack--is-stacked mm-component-stack--has-footer'}):
    for j in i('a'):
        link1 = "https://fashionista.com/fashion-week" + j.get('href')
        link.append(link1)
        print(link1)
```

```python
set_link = set(link)
link = list(set_link)
len(link)
```

```python
# only article content is fetched
csv_file = open('Fashionista.csv', 'w',encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Article Content'])
count = 0
for i in link:
    driver.get(i)
    soup = BeautifulSoup(driver.page_source)
    content = ""
    for i in soup.findAll('p'):
        content1= i.text
        content = content +" "+ content1
    csv_writer.writerow([content])
    print(count)
    count= count + 1
```
